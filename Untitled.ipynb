```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.model_selection import cross_val_score
from sklearn.utils import column_or_1d
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, explained_variance_score
from sklearn.utils import check_array

import seaborn as sns

from sklearn.ensemble import GradientBoostingRegressor
from sklearn.svm import SVR
```

```python
youtube_videos = pd.read_csv('youtube_videos.tsv',sep='\t')
transcoding_measurements = pd.read_csv('transcoding_mesurment.tsv',sep='\t')
```

```python
X, y = np.split(transcoding_measurements,[-1],axis=1)
y = column_or_1d(y)
```

```python
to_drop = ['id','umem','duration','frames','size']
```

```python
X = X.drop(columns=to_drop)
X
```

```python
X = pd.get_dummies(X,columns=['codec'])
X = pd.get_dummies(X,columns=['o_codec'])
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(X)
X = pd.DataFrame(x_scaled)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
reg = GradientBoostingRegressor(n_estimators=300, learning_rate=0.1, max_depth=8, random_state=0, loss='ls')
```

```python
reg.fit(X_train,y_train)
```

```python
pred = reg.predict(X_test)
```

```python
mean_squared_error(y_test,pred)
```

```python
mean_absolute_error(y_test,pred)
```

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
mean_absolute_percentage_error(y_test,pred)
```

```python
explained_variance_score(y_test,pred)
```

```python
plt.plot(y_test, color="blue")
plt.plot(pred, color="red")
plt.plot()
```

```python
params = {'n_estimators': 300, 'max_depth': 8, 'min_samples_split': 2,
          'learning_rate': 0.01, 'loss': 'ls'}

test_score = np.zeros((params['n_estimators'],), dtype=np.float64)

for i, pred in enumerate(reg.staged_predict(X_test)):
    test_score[i] = reg.loss_(y_test, pred)
    
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title('Desvio')
plt.plot(np.arange(params['n_estimators']) + 1, reg.train_score_, 'b-',
         label='Training Set Deviance')
plt.plot(np.arange(params['n_estimators']) + 1, test_score, 'r-',
         label='Test Set Deviance')
plt.legend(loc='upper right')
plt.xlabel('Iteracoes')
plt.ylabel('Desvio')
```
